```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#Player Count
#calcaulating unique players based on SN 
player_count = len(purchase_data["SN"].unique())

#Creating a data frame for only Total players and displaying it
total_player_count = pd.DataFrame([{"Total Player" : player_count}])
total_player_count
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Purchaing Analysis (Total)
#calculating unique items, average price, number of total purchases and total revenuew
unique_items = len(purchase_data["Item ID"].unique())
average_price = purchase_data["Price"].mean()
num_purchases = purchase_data["Purchase ID"].count()
total_revenue = purchase_data["Price"].sum()

#Create a new data frame and assign each value
purchasing_analysis = pd.DataFrame([{"Number of Unique Item" : unique_items, 
                                        "Average Price":'${:,.2f}'.format(average_price), 
                                        "Number of Purchases":num_purchases,
                                        "Total Revenue": '${:,.2f}'.format(total_revenue)}])

#rearranging the order of column
purchasing_analysis[["Number of Unique Item", "Average Price", "Number of Purchases", "Total Revenue"]]
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Gender Demographics
#creating a new data frame dropping all duplicates from origianl data frame
unique_players_count = purchase_data.drop_duplicates(["SN"], keep = "last")

#from the above created non-duplicate data frame create a new data frame based on 'Gender'
gender_demographics = pd.DataFrame(unique_players_count["Gender"].value_counts())

#calcuating each players per gender
total_count = unique_players_count["Gender"].count()

#Calculating percentage of total players per gender for unique players 
gender_percent = unique_players_count["Gender"].value_counts() *100 / total_count 

#Assigning the values, renamaing the column, displaying the data frame
gender_demographics["Percentage of Player"] = gender_percent.map('{:,.2f}%'.format)
gender_demographics = gender_demographics.rename(columns = {"Gender":"Total Count"})
gender_demographics
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Purchasing Analysis (Gender)
# creating a new data frame based on original data frame by grouping the data on Gender
purchase_analysis_data = purchase_data.groupby(["Gender"])

#per each group in gender, calculating the total count, average and total sum
purchase_count = purchase_analysis_data["Purchase ID"].count()
average_purchase_price = purchase_analysis_data["Price"].mean()
total_purchase_price = purchase_analysis_data["Price"].sum()

#calcualte the average total purchase based on unique members in each gender
average_total_purchases_per_person = total_purchase_price / unique_players_count["Gender"].value_counts()

#Creating a new data frame based on grouped dataframe and doing the assignments, formating the data
purchase_analysis = pd.DataFrame(purchase_analysis_data["Purchase ID"].count())
purchase_analysis["Average Purchase Price"] = average_purchase_price.map("${:,.2f}".format)
purchase_analysis["Total Purchase Value"] = total_purchase_price.map("${:,.2f}".format)
purchase_analysis["Avg Total Purchase per Person"] = average_total_purchases_per_person.map("${:,.2f}".format)

#Dislplay the data frame
purchase_analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Age Demographics
#creating new data frame dropping the duplicates
age_unique_data = purchase_data.drop_duplicates(["SN"], keep = "last")

#count total players
players_count = age_unique_data["Age"].count()

#Establish bings for ages and assign the labels
bins = [0,9,14,19,24,29,34,39,100]
age_group_labels = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

#Based on bins, create a new column by assigning the age group values based on bins
age_unique_data ["Age Group"] = pd.cut(age_unique_data["Age"], bins, labels =age_group_labels)

#create a new data frame by grouping based on age group
age_demographics = age_unique_data.groupby("Age Group")

#create a new data frame based on above grouped data frame 
age_demographics_summary = pd.DataFrame(age_demographics.count())

#assign the values
age_demographics_summary["Percentage of Players"] = (age_demographics_summary["Purchase ID"] *100 / players_count).map("{:,.2f}".format)
age_demographics_summary = age_demographics_summary.rename(columns = {"Purchase ID":"Total Count"})

#diaplay the resutls for required columns
age_demographics_summary[["Total Count","Percentage of Players"]]
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Purchasing Analysis
#create a new data frame based on original data frame
purchase_a_data = purchase_data

#create bins and set the labels
bins = [0,9,14,19,24,29,34,39,100]
age_labels = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

#create a set new data column based on Age in each bin
purchase_a_data ["Age Group"] = pd.cut(purchase_a_data["Age"], bins, labels =age_labels)

#create a new data base with group by age grouo
purchase_a = purchase_a_data.groupby("Age Group")

#calcualte the average and total sum
price_mean = purchase_a["Price"].mean()
price_sum = purchase_a["Price"].sum()

#Create a new data frame dropping the duplicates to calculate unique purchasers based on SN
purchase_person = pd.DataFrame(purchase_data.drop_duplicates("SN", keep = "last"))
purchase_person = purchase_person.groupby('Age Group')
person_age_group = purchase_person['SN'].count()

#Calacuate the average purchase per person for unique purchasers per age group
price_per_person = price_sum / person_age_group

#creating new data frame to set and assign the values calculated
purchase_a_summary = pd.DataFrame(purchase_a["Purchase ID"].count())
purchase_a_summary["Average Purchase Price"] = price_mean.map("${:,.2f}".format)
purchase_a_summary["Total Purchase Value"] = price_sum.map("${:,.2f}".format)
purchase_a_summary["Avg Total Purchase per Person"] = price_per_person.map("${:,.2f}".format)

#rename the column
purchase_a_summary = purchase_a_summary.rename(columns = {"Purchase ID":"Purchase Count"})

#display the specific data frame
purchase_a_summary[["Purchase Count","Average Purchase Price","Total Purchase Value", "Avg Total Purchase per Person"]]
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Top Spenders
#create a new data frame based on original data frame
top_spender_data = pd.DataFrame(purchase_data)

#group new data frame based on SN
top_spender = top_spender_data.groupby("SN")

#Calcuate average and total sum of the price per group in SN
spender_price_mean = top_spender["Price"].mean()
spender_price_sum = top_spender["Price"].sum()

#create new data frame to assign calculated values
spender_summary = pd.DataFrame(top_spender["Purchase ID"].count())
spender_summary["Average Purchase Price"] = spender_price_mean.map("${:,.2f}".format)
spender_summary["Total Purchase Value"] = spender_price_sum.map("${:,.2f}".format)

#rename the column
spender_summary = spender_summary.rename(columns = {"Purchase ID":"Purchase Count"})

#display the required columns
spender_summary[["Purchase Count","Average Purchase Price","Total Purchase Value"]] 

#sort the data based on purchase count
spender_summary.sort_values("Purchase Count",ascending=False).head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Most popular items
#create a new data frame based on original data frame
popular_data = purchase_data

#group up the new data frame on 2 columns
popular_item_data = popular_data.groupby(["Item ID", "Item Name"])

#calculate the average and total sum of price
item_price = popular_item_data["Price"].mean()
spender_price_sum = popular_item_data["Price"].sum()

#create new data frame from the grouped data frame
item_summary = pd.DataFrame(popular_item_data["Purchase ID"].count())

#Format the values 
item_summary["Item Price"] = item_price.map("${:,.2f}".format)
item_summary["Total Purchase Value"] = spender_price_sum.map("${:,.2f}".format)

#rename the columns
item_summary = item_summary.rename(columns = {"Purchase ID":"Purchase Count"})

#display the data and sorting descending order based on purchase counts per SN
item_summary[["Purchase Count","Item Price","Total Purchase Value"]] 
item_summary.sort_values("Purchase Count",ascending=False).head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
#Most Profitable Items
#re-calculate the sum of total purchases of above data frame grouped on item id and item name
#this is used to retain the sum values in numeric for sorting purpose
#in above table, this value is stored as string
profitable_items_sum = popular_item_data["Price"].sum()

#assign it to same data frame 
item_summary["Total Purchase Value"] = profitable_items_sum  
 
#sort the table in descending order on total purchase value
most_profitable_summary = item_summary.sort_values("Total Purchase Value",ascending=False)

#display the data frame to confirm descending order
most_profitable_summary.head()

#format the column 
profitable_items = profitable_items_sum.map("${:,.2f}".format)

#assign this formatted value
most_profitable_summary["Total Purchase Value"] = profitable_items

#dislay preview
most_profitable_summary.head()
```
